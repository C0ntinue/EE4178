### 0. Define Hyper-parameters and Device Configuration

```python
!pip install torchtext
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchsummary
import torchtext
from torchtext import data, datasets
```

```python
# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyper-parameters
batch_size = 64
learning_rate = 0.005
num_epochs = 5

out_node = 1 ## binary {neg, pos}
num_layers = 3
embedd_dim = 512
hidden_dim = 256
valid_percent = 0.2
MIN_FREQ = 100
```

### 1. Load Data

#### 1.0 Download IMDB

```python
TEXT = data.Field(sequential=True, batch_first=True, lower=True)
LABEL = data.LabelField(sequential=False, dtype=torch.float, batch_first=True)

train_data, test_data = datasets.IMDB.splits(TEXT, LABEL)
```

```python
# cf) check data
print("train:", len(train_data), "test: ",len(test_data))
print(vars(train_data.examples[0]))
```

#### 1.1 Make Validation Dataset
Since IMDB dataset only has train/test splits without validation, we should make valiation set.
It is available by .split() method.

IMDB 데이터셋에서는 valid set을 따로 제공하지 않음 > 직접 만들기


```python
train_data, valid_data = train_data.split(split_ratio=1-valid_percent)
```

```python
# cf) check each data
print("train: ", len(train_data))
print("valid: ", len(valid_data))
print("test : ", len(test_data))
```

#### 1.2 Build Vocabulary Set
단어 집합 만들기

```python
TEXT.build_vocab(train_data, min_freq=MIN_FREQ) # N번 이상 나온 단어들에 대하여 vocab set 생성
LABEL.build_vocab(train_data)
```

```python
# cf) check Vocab Set
vocab_size = len(TEXT.vocab)
print("단어 집합의 크기: ", vocab_size)
```

```python
# cf) vocab set을 단어 빈도 수 기준이 아닌, vocab set size를 정해두고 만들 수 있음
# ex) TEXT.build_vocab(train_data, max_size=20000) 는 len(TEXT.vocab)이 20000인 vocab set
#     즉, 20000개의 vocabulary로 이루어진 vocab set
```

```python
# cf) 가장 빈도수가 높은 20개 vocab
TEXT.vocab.itos[:20]
```

###2. Define Dataloader

```python
train_loader, valid_loader = data.BucketIterator.splits(
    (train_data, valid_data),
    batch_size=batch_size,
    device=device
)

test_loader = data.BucketIterator(dataset=test_data,
                                  batch_size=batch_size,
                                  train=False,
                                  device=device)
```

```python
vocab_size
```

### 3. Build Model

```python
class Sentiment(nn.Module):
  def __init__(self, num_vocab, embedding_dim, hidden_dim, num_layers, out_node, model="RNN"):
    super(Sentiment, self).__init__()

    self.model = model
    self.num_vocab = num_vocab
    
    self.embed = nn.Embedding(num_embeddings=num_vocab, embedding_dim=embedd_dim)
    self.rnn = nn.RNN(input_size=embedding_dim, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True)
    self.lstm = nn.LSTM(input_size=embedding_dim, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True)
    self.gru = nn.GRU(input_size=embedding_dim, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True)

    self.fc = nn.Linear(hidden_dim, out_node)
    self.sigmoid = nn.Sigmoid()


  def forward(self, x):
    x = self.embed(x)

    if self.model == "RNN":
      out, hidden = self.rnn(x)
    elif self.model == "LSTM":
      out, (hidden, cell_state) = self.lstm(x)
    elif self.model == "GRU":
      out, hidden = self.gru(x)
    else:
      print("choose a model in ['RNN', 'LSTM', 'GRU']")
      raise

    out = self.fc(out[:,-1,:])

    return out
```

```python
model = Sentiment(vocab_size, embedd_dim, hidden_dim, num_layers, out_node, "RNN").to(device)
```

### 4. Set Loss & Optimizer

```python
criterion = nn.BCELoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

### 5. Train & Test

#### 5.0 Train funtion

```python
def train(model, train_loader, optimizer, criterion):
  epoch_loss, acc = 0, 0
  model.train()

  for i, data in enumerate(train_loader):
    text = data.text
    label = data.label

    out = model(text).squeeze(1)
    loss = criterion(out, label)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    acc += (torch.round(out)==label).sum()
    epoch_loss += loss.item()
    

  return epoch_loss/(i+1), acc / len(train_loader.dataset) * 100
```

#### 5.1 Evaluation function

```python
def evaluate(model, data_loader, criterion):
  epoch_loss, acc = 0, 0
  model.eval()

  with torch.no_grad():
    for data in data_loader:
      text = data.text
      label = data.label

      out = model(text).squeeze(1)
      loss = criterion(out, label)

      acc += (torch.round(out)==label).sum()
      epoch_loss += loss.item()

  return epoch_loss/len(data_loader), acc / len(data_loader.dataset) * 100
```

#### 5.2 Train

```python
best_valid_loss = float('inf')
best_epoch = 0
for epoch in range(num_epochs):
  train_loss, train_acc = train(model, train_loader, optimizer, criterion)
  valid_loss, valid_acc = evaluate(model, valid_loader, criterion)


  if valid_loss < best_valid_loss:
    best_valid_loss = valid_loss
    best_epoch = epoch
    torch.save(model.state_dict(), "epoch_{}.pth".format(epoch))

  print("Epoch[{}/{}], Train Loss:{:.4f}, Train Acc:{:.2f}, Valid Loss:{:.4f}, Valid Acc:{:.2f}".format(epoch+1, num_epochs, train_loss, train_acc, valid_loss, valid_acc))
```

#### 5.3 Test

```python
model.load_state_dict(torch.load('./epoch_{}.pth'.format(best_epoch)))

RNN_loss, RNN_acc = evaluate(model, test_loader, criterion)
print("RNN's IMDB Accuracy:{:.2f}% (Loss:{:.4f})".format(RNN_acc, RNN_loss))
```

### 6. Other models: LSTM, GRU

####6.1 LSTM

```python
model = Sentiment(len(TEXT.vocab), embedd_dim, hidden_dim, num_layers, out_node, "LSTM").to(device)

criterion = nn.BCEWithLogitsLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# LSTM #
# Train
print("____LSTM____")
best_valid_loss = float('inf')
best_epoch = 0
for epoch in range(num_epochs):
  train_loss, train_acc = train(model, train_loader, optimizer, criterion)
  valid_loss, valid_acc = evaluate(model, valid_loader, criterion)


  if valid_loss < best_valid_loss:
    best_valid_loss = valid_loss
    best_epoch = epoch
    torch.save(model.state_dict(), "LSTM_epoch_{}.pth".format(epoch))

  print("Epoch[{}/{}], Train Loss:{:.4f}, Train Acc:{:.2f}, Valid Loss:{:.4f}, Valid Acc:{:.2f}".format(epoch+1, num_epochs, train_loss, train_acc, valid_loss, valid_acc))

# Test
model.load_state_dict(torch.load('./LSTM_epoch_{}.pth'.format(best_epoch)))
LSTM_loss, LSTM_acc = evaluate(model, test_loader, criterion)
print("\nLSTM's IMDB Accuracy:{:.2f}% (Loss:{:.4f})".format(LSTM_acc, LSTM_loss))
```

####6.2 GRU

```python
model = Sentiment(len(TEXT.vocab), embedd_dim, hidden_dim, num_layers, out_node, "GRU").to(device)

criterion = nn.BCEWithLogitsLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# GRU #
# Train
print("____GRU____")
best_valid_loss = float('inf')
best_epoch = 0
for epoch in range(num_epochs):
  train_loss, train_acc = train(model, train_loader, optimizer, criterion)
  valid_loss, valid_acc = evaluate(model, valid_loader, criterion)


  if valid_loss < best_valid_loss:
    best_valid_loss = valid_loss
    torch.save(model.state_dict(), "GRU_epoch_{}.pth".format(epoch))

  print("Epoch[{}/{}], Train Loss:{:.4f}, Train Acc:{:.2f}, Valid Loss:{:.4f}, Valid Acc:{:.2f}".format(epoch+1, num_epochs, train_loss, train_acc, valid_loss, valid_acc))

# Test
model.load_state_dict(torch.load('./GRU_epoch_{}.pth'.format(best_epoch)))
GRU_loss, GRU_acc = evaluate(model, test_loader, criterion)
print("\nGRU's IMDB Accuracy:{:.2f}% (Loss:{:.4f})".format(GRU_acc, GRU_loss))
```
